```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import tweepy
import json
#import time
import pandas as pd
import keys
```

```python
consumer_key = keys.consumer_key

consumer_secret = keys.consumer_secret

access_token = keys.access_token

access_token_secret = keys.access_token_secret
```

```python
def auth(consumer_key, consumer_secret,access_token, access_token_secret):
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth, wait_on_rate_limit = True)
    return api

def user(api, name):
    return api.get_user(screen_name=name)

def user_id(api,name_id):
    return api.get_user(user_id=name_id)

def following_id(api, name_id):
    print("name_id",name_id)
    return api.get_friend_ids(user_id=name_id)
```

```python
name = "sarvesh_shroff"
api = auth(consumer_key, consumer_secret,access_token, access_token_secret)
user_detail = user(api,name)
```

```python
list_following = following_id(api,user_detail.id)
list_following.append(user_detail.id)
print(list_following)
number_of_following = len(list_following)
print(number_of_following)
df=pd.DataFrame(list_following,columns=['label'])
df.index.name = 'id'
df.to_csv("node.csv")
```

```python
adj=np.empty(2,dtype=int)
adj
```

```python
for i in range(number_of_following):
    follow = following_id(api,list_following[i])
    for j in range(number_of_following):
        for k in follow:
            if list_following[j] == k:
                c=[i,list_following.index(k)]
                adj=np.vstack((adj,c))
```

```python
df=pd.DataFrame(adj[1:],columns=['Source', 'Target'])
df.to_csv("edge.csv",index=False)
```

### Visualization using Networkx

```python
G = nx.DiGraph()
for i,j in adj[1:]: 
    G.add_edge(i,j)
nx.draw(G) 
plt.show() 
```
